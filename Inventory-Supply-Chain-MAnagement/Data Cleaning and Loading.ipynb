# Loading all the data in the superstore database in mysql

```python
host='127.0.0.1'
user='root'
port=111111
db='superstore'
passwd='**********'
```

```python
engine=create_engine(f'mysql+pymysql://{user}:{passwd}@{host}/{db}')
engine
```

```python
tables = {
    'products': products,
    'customer_details': customer_details,
    'brand_products': brand_products,
    'transactions': transcations,  # Fixed spelling if it's 'transactions'
    'categories': categories,
    'cashier': cashier
}

for table_name, df in tables.items():
    try:
        df.to_sql(table_name, engine, if_exists='replace', index=False)
        print(f"Successfully inserted into table: {table_name}")
    except Exception as e:
        print(f"Error inserting into table: {table_name} - {e}")
```
